```
import requests 
import re
from bs4 import BeautifulSoup

url = 'https://en.wikipedia.org/wiki/List_of_U.S._states_and_territories_by_GDP_per_capita'
page = requests.get(url).text #query the website and return the html to the variable 'page'
#page = requests.get(url).content
soup = BeautifulSoup(re.sub("<!--|-->","", page), "lxml")
#soup = BeautifulSoup(page, 'html.parser') #using beautiful soup to parse the html and store in variable'soup'
print(soup.prettify())

#r = requests.get('https://www.kbb.com/vehicles/hub/_costtoown/?vehicleid=415933')
#html_soup = BeautifulSoup(r.text, 'html.parser')
#Insurance = html_soup.find('div',string="Insurance").find_next().text
#print(Insurance)
```

```
#my_table = soup.find('table',{'class':'wikitable sortable jquery-tablesorter'})
my_table = soup.findAll('table')[0]
print(my_table)
```

```
links = my_table.find_all('th')
print(links)
```

```
#Extract headings from the links 
Texts = []
for link in links:
  Texts.append(link.get_text(strip=True))

headings = Texts[0:10]
print(headings)
```

```
#Find all the rows of the table, organized by the headings
table_rows = my_table.find_all('tr')
rows = []
for tr in table_rows: 
  td = tr.find_all('td')
  row = [tr.get_text(strip=True) for tr in td]
  rows.append(row)

print(rows)
```

```
import pandas as pd

df = pd.DataFrame(rows, columns = headings)
df = df.drop(df.index[0])
print(df)
```

```
df.to_csv('US_States_GDP.csv',index = False, header = True)
```
