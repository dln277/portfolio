**1. Import Data**

```
import pandas as pd 

X_train = pd.read_csv('Training_values.csv')
y_train = pd.read_csv('Training_labels.csv')
X_test = pd.read_csv('Test_set_values.csv')
```

```
X_train.info()
```

```
X_train.head()
```

**2. Data Processing**

```
#Dropping unneccesary columns 

X_train = X_train.drop(columns=['date_recorded','wpt_name','num_private','longitude','latitude','subvillage',
                                'region','lga','ward','recorded_by','scheme_name','management','payment','water_quality',
                                'quantity_group','source','source_class','waterpoint_type_group','district_code','region_code',
                                'extraction_type','extraction_type_group','basin'])

X_test = X_test.drop(columns=['date_recorded','wpt_name','num_private','longitude','latitude','subvillage',
                                'region','lga','ward','recorded_by','scheme_name','management','payment','water_quality',
                                'quantity_group','source','source_class','waterpoint_type_group','district_code','region_code',
                              'extraction_type','extraction_type_group','basin'])
```

```
X_train.info()
```

```
#Transform meta columns to binary columns- 1 if contains value, 0 if NaN

X_train['funder'] = X_train['funder'].notnull().astype('int')
X_train['funder'] = X_train['funder'].fillna(0)
X_train['funder'] = X_train['funder'].replace('Not Known',0)

X_train['installer'] = X_train['installer'].notnull().astype('int')
X_train['installer'] = X_train['installer'].fillna(0)
X_train['installer'] = X_train['installer'].replace('Not Known',0)
```

```
#Transform categorical binary columns to take binary values 
#Fill NaN with 0

X_train['public_meeting'] = X_train['public_meeting'].fillna(0)
X_train['permit'] = X_train['permit'].fillna(0)
X_train[['public_meeting','permit']] = X_train[['public_meeting','permit']].astype(int)
```

```
#Transform categorical columns to binary values with One Hot Encoder 

X_train['scheme_management'] = X_train['scheme_management'].fillna('None')

X_train[['scheme_management','extraction_type_class',
              'management_group','payment_type','quality_group',
              'quantity','source_type','waterpoint_type']] = X_train[['scheme_management','extraction_type_class',
              'management_group','payment_type','quality_group',
              'quantity','source_type','waterpoint_type']].astype('category')
X_train = pd.get_dummies(X_train)
X_train.head()
```

```
#Preprocess numerical variables, impute NaN with mean

X_train['construction_year'] = X_train['construction_year'].replace(0,X_train['construction_year'].mean())
```

```
X_train = X_train.drop(columns='id')
y_train = y_train.drop(columns='id')
```

**3. Train & Fit Models**

```
from sklearn.tree import DecisionTreeClassifier
tree = DecisionTreeClassifier(random_state=0)
tree.fit(X_train,y_train)
```

```
from sklearn.naive_bayes import GaussianNB
nb = GaussianNB()
nb.fit(X_train,y_train)
```

```
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier()
knn.fit(X_train,y_train)
```

```
from sklearn.ensemble import RandomForestClassifier
forest = RandomForestClassifier()
forest.fit(X_train,y_train)
```

```
from sklearn.ensemble import GradientBoostingClassifier
gradient = GradientBoostingClassifier()
gradient.fit(X_train,y_train)
```

```
from sklearn.svm import LinearSVC #(setting multi_class=”ovr”)
svc = LinearSVC(multi_class='ovr')
svc.fit(X_train,y_train)
```

```
from sklearn.linear_model import LogisticRegression #(setting multi_class=”ovr”)
lr = LogisticRegression(multi_class='ovr')
lr.fit(X_train,y_train)
```

```
from sklearn.linear_model import SGDClassifier
sgd = SGDClassifier()
sgd.fit(X_train,y_train)
```

```
from sklearn.model_selection import RepeatedStratifiedKFold
from sklearn.model_selection import cross_val_score

def baseline_model(model_name):
  model = model_name
  cv = RepeatedStratifiedKFold(n_splits=10, n_repeats=3, random_state=1)
  scores = cross_val_score(model,X_train,y_train,cv=cv,scoring='accuracy')
  print(model,'Accuracy: %.3f' % (scores.mean()))

baseline_model(tree)
baseline_model(nb)
baseline_model(knn)
baseline_model(forest)
baseline_model(gradient)
baseline_model(svc)
baseline_model(lr)
baseline_model(sgd)
```

```
#Understand which features are the most important (remove those that are not to improve performance)
import numpy as np 

importances = pd.DataFrame({'feature':X_train.columns,'importance':np.round(tree.feature_importances_,3)})
importances = importances.sort_values('importance',ascending=False).set_index('feature')
importances.head(10)
```

```
X_train = X_train.filter(['gps_height','quantity_dry','population','construction_year','waterpoint_type_other','amount_tsh'])
```

```
from sklearn.model_selection import cross_val_score 

scores_knn = cross_val_score(knn,X_train,y_train,cv=3,scoring='accuracy')
scores_tree = cross_val_score(tree,X_train,y_train,cv=3,scoring='accuracy')
scores_forest = cross_val_score(forest,X_train,y_train,cv=3,scoring='accuracy')
scores_gradient = cross_val_score(gradient,X_train,y_train,cv=3, scoring='accuracy')
scores_lr = cross_val_score(lr,X_train,y_train,cv=3, scoring='accuracy')

cross_val = pd.DataFrame({
        'Model': ['KNN Classifier','Decision Tree','Random Forest','Gradient Boosting','Logistic Regression'],
        'Score': [scores_knn,scores_tree,scores_forest,scores_gradient,scores_lr],
        'Mean': [scores_knn.mean(),scores_tree.mean(),scores_forest.mean(),scores_gradient.mean(),scores_lr.mean()],
        'Standard Deviation': [scores_knn.std(),scores_tree.std(),scores_forest.std(),scores_gradient.std(),scores_lr.std()]})

cross_val_df = cross_val.sort_values(by='Mean',ascending=False)
cross_val_df = cross_val_df.set_index('Mean')

cross_val_df.head()
```

```
#Hyperparameter tuning to improve random forest's performance

from sklearn.model_selection import GridSearchCV

param_grid = { "criterion" : ["gini", "entropy"],
              "min_samples_leaf" : [1, 5, 10, 25, 50, 70],
              "min_samples_split" : [2, 4, 10, 12, 16, 18, 25, 35],
              "n_estimators": [100, 400, 700, 1000, 1500]}

rf = RandomForestClassifier(n_estimators=100, max_features='auto', oob_score=True, random_state=1, n_jobs=-1)

clf = GridSearchCV(estimator=rf, param_grid=param_grid, n_jobs=-1)
clf.fit(X_train, y_train)

clf.best_params_
```

```
#Preprocessing X_test 

X_test[['scheme_management','extraction_type_class',
              'management_group','payment_type','quality_group',
              'quantity','source_type','waterpoint_type']] = X_test[['scheme_management','extraction_type_class',
              'management_group','payment_type','quality_group',
              'quantity','source_type','waterpoint_type']].astype('category')
X_test = pd.get_dummies(X_test)
X_test['construction_year'] = X_test['construction_year'].replace(0,X_test['construction_year'].mean())
X_test = X_test.filter(['gps_height','quantity_dry','population','construction_year','waterpoint_type_other','amount_tsh'])
X_test.info()
```

```
rf = RandomForestClassifier(n_estimators=100, max_features='auto', oob_score=True, random_state=1, n_jobs=-1)
rf.fit(X_train,y_train)
y_pred = rf.predict(X_test)
print(y_pred)
```
